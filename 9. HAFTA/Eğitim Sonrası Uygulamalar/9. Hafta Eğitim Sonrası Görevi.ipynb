# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda:
    def __init__(self):
        self.tvAcik = False
        self.sesSeviyesi = 0
        self.kanallar = []
        self.aktifKanal = None

    def tvAc(self):
        if not self.tvAcik:
            self.tvAcik = True
            print("Televizyon açıldı.")
        else:
            print("Televizyon zaten açık.")

    def tvKapat(self):
        if self.tvAcik:
            self.tvAcik = False
            self.aktifKanal = None
            print("Televizyon kapatıldı.")
        else:
            print("Televizyon zaten kapalı.")

    def sesAyarlari(self):
        if not self.tvAcik:
            print("Televizyon kapalı. Ses ayarlarını değiştiremiyorsunuz.")
            return
        while True:
            try:
                yeniSes = int(input(f"Mevcut ses seviyesi: {self.sesSeviyesi}. Yeni ses seviyesini girin (0-100): "))
                if 0 <= yeniSes <= 100:
                    self.sesSeviyesi = yeniSes
                    print(f"Ses seviyesi {self.sesSeviyesi} olarak ayarlandı.")
                    break
                else:
                    print("Ses seviyesi 0 ile 100 arasında olmalıdır.")
            except ValueError:
                print("Geçerli bir sayı girin.")

    def kanalEkle(self):
        if not self.tvAcik:
            print("Televizyon kapalı. Kanal ekleyemezsiniz.")
            return
        kanal = input("Eklemek istediğiniz kanalın adını girin: ")
        if kanal not in self.kanallar:
            self.kanallar.append(kanal)
            print(f"{kanal} kanalı eklendi.")
        else:
            print("Bu kanal zaten mevcut.")

    def kanalSayisiniOgrenme(self):
        print(f"Toplam kanal sayısı: {len(self.kanallar)}")

    def rastgeleKanalaGecme(self):
        if not self.tvAcik:
            print("Televizyon kapalı. Kanal değişikliği yapamazsınız.")
            return
        if not self.kanallar:
            print("Eklenmiş kanal yok.")
            return
        self.aktifKanal = random.choice(self.kanallar)
        print(f"Şu anki kanal: {self.aktifKanal}")

    def tvBilgileri(self):
        if not self.tvAcik:
            print("Televizyon kapalı. Bilgileri göremiyorsunuz.")
            return
        aktifKanal = self.aktifKanal if self.aktifKanal else "Hiçbir kanal seçilmedi"
        print(f"Televizyon Açık: {self.tvAcik}")
        print(f"Ses Seviyesi: {self.sesSeviyesi}")
        print(f"Kanallar: {', '.join(self.kanallar) if self.kanallar else 'Hiç kanal yok.'}")
        print(f"Aktif Kanal: {aktifKanal}")

def main():
    kumanda = Kumanda()
    while True:
        print("\nYapmak istediğiniz işlemi seçin:")
        print("1. TV Aç")
        print("2. TV Kapat")
        print("3. Ses Ayarları")
        print("4. Kanal Ekle")
        print("5. Kanal Sayısını Öğrenme")
        print("6. Rastgele Kanala Geçme")
        print("7. Televizyon Bilgileri")
        print("q. Çıkış")
        
        secim = input("Seçiminizi yapın: ").strip().lower()
        
        if secim == '1':
            kumanda.tvAc()
        elif secim == '2':
            kumanda.tvKapat()
        elif secim == '3':
            kumanda.sesAyarlari()
        elif secim == '4':
            kumanda.kanalEkle()
        elif secim == '5':
            kumanda.kanalSayisiniOgrenme()
        elif secim == '6':
            kumanda.rastgeleKanalaGecme()
        elif secim == '7':
            kumanda.tvBilgileri()
        elif secim == 'q':
            print("Programdan çıkılıyor...")
            time.sleep(1)
            break
        else:
            print("Geçersiz seçim. Lütfen tekrar deneyin.")

if __name__ == "__main__":
    main()
```

# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
def vkiHesapla(boy, kilo):
    return kilo / (boy ** 2) # VKİ hesaplama formülü: kilo / (boy * boy)

def vkiSiniflandir(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 24.9:
        return "Normal kilolu"
    elif 25 <= vki < 29.9:
        return "Fazla kilolu"
    elif 30 <= vki < 39.9:
        return "Obez"
    else:
        return "İleri derecede obez (morbid obez)"

def veriGirisiVeKaydet():
    girisSayisi = int(input("Kaç giriş yapacaksınız? "))
    with open("veriKayit.txt", "w") as dosya:
        for _ in range(girisSayisi):
            isim = input("İsim: ")
            soyisim = input("Soyisim: ")
            yas = int(input("Yaş: "))
            boy = float(input("Boy (metre cinsinden): "))
            kilo = float(input("Kilo (kg cinsinden): "))

            vki = vkiHesapla(boy, kilo)
            sinif = vkiSiniflandir(vki)

            kayit = f"İsim: {isim}, Soyisim: {soyisim}, Yaş: {yas}, Boy: {boy}, Kilo: {kilo}, VKİ: {vki:.2f}, Sınıf: {sinif}\n"
            dosya.write(kayit)
            print(kayit)

# Programı çalıştır
veriGirisiVeKaydet()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import datetime
menu = {
    "Muz": 5.0,
    "Çilek": 7.0,
    "Ananas": 8.0,
    "Ispanak": 4.0,
    "Badem Sütü": 6.0,
    "Yulaf": 3.0
}

def menuGoruntule():
    print("\n--- YetBar Menü ---")
    for urun, fiyat in menu.items():
        print(f"{urun}: {fiyat} TL")

def satinAl():
    smoothie = []
    toplamFiyat = 0.0

    while True:
        urun = input("\nEklemek istediğiniz ürünü giriniz (çıkmak için 'q'): ")
        if urun.lower() == 'q':
            break
        elif urun in menu:
            smoothie.append(urun)
            toplamFiyat += menu[urun]
            print(f"{urun} eklendi. Şu ana kadar toplam: {toplamFiyat} TL")
        else:
            print("Geçersiz ürün. Lütfen menüden bir ürün seçin.")

    if smoothie:
        kayit = smoothieKaydet(smoothie, toplamFiyat)
        print(f"\nSmoothie siparişiniz kaydedildi. Sipariş detayları: {kayit}")
    else:
        print("Sipariş vermediniz.")

def smoothieKaydet(smoothie, toplamFiyat):
    tarih = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    kayit = f"Tarih: {tarih}, Ürünler: {', '.join(smoothie)}, Toplam Fiyat: {toplamFiyat} TL\n"

    with open("smoothieKayitlari.txt", "a") as dosya:
        dosya.write(kayit)

    return kayit

def anaMenu():
    while True:
        print("\nYetBar'a hoş geldiniz! Yapmak istediğiniz işlemi seçin:")
        print("1. Menüyü görüntüleme")
        print("2. Smoothie satın alma")
        print("3. Çıkış")

        secim = input("Seçiminiz: ")

        if secim == "1":
            menuGoruntule()
        elif secim == "2":
            satinAl()
        elif secim == "3":
            print("Çıkış yapılıyor...")
            break
        else:
            print("Geçersiz seçim. Lütfen tekrar deneyin.")

anaMenu()
```

# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
def urunBilgisiKaydet():
    urunAdi = input("Ürün adını girin: ")
    urunFiyati = input("Ürün fiyatını girin: ")

    with open("urunler.txt", "a") as dosya:
        dosya.write(f"{urunAdi}, {urunFiyati}\n")
    
    print(f"{urunAdi} ürünü başarıyla kaydedildi.")

urunBilgisiKaydet()
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız
def kelimeDegistir(dosyaAdi, eskiKelime, yeniKelime):
    with open(dosyaAdi, "r") as dosya:
        icerik = dosya.read()
    
    yeniIcerik = icerik.replace(eskiKelime, yeniKelime)

    with open(dosyaAdi, "w") as dosya:
        dosya.write(yeniIcerik)
    
    print(f"{eskiKelime} kelimesi {yeniKelime} ile değiştirildi.")

eskiKelime = input("Değiştirmek istediğiniz kelimeyi girin: ")
yeniKelime = input("Yeni kelimeyi girin: ")

kelimeDegistir("urunler.txt", eskiKelime, yeniKelime)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Arac:
    def __init__(self, ad, hiz, kilometre):
        self.ad = ad
        self.hiz = hiz
        self.kilometre = kilometre

    def bilgiGoster(self):
        print(f"Ad: {self.ad} Hız: {self.hiz} Kilometre: {self.kilometre}")

class Otobus(Arac):
    pass

otobus = Otobus("School Volvo", 180, 12)
otobus.bilgiGoster()

````
